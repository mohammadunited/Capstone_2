#### The  delimiter = '|' seprates the code

```python
upper_list = pd.read_csv('customer_list_updated.csv', delimiter = '|')
columns = ('cust_id', 'date', 'time', 'name', 'email', 'phone', 'sms-opt-out')
```

#### Convert the value to a string to use replace and replace the special charecters 

```python
for i in upper_list['name']: 
    # Convert the value to a string to use replace
     name = str(['name'])  
    #the code below is getting rid of the extra charecters
     if '&' in name:
         upper_list.at[i, 'name'] = name.replace('&','')  
     elif '^' in name:
         upper_list.at[i, 'name'] = name.replace('^','')  
     elif '.' in name:
         upper_list.at[i, 'name'] = name.replace('.','')  
     elif '-' in name:
         upper_list.at[i, 'name'] = name.replace('-','') 
     elif '!' in name:
         upper_list.at[i, 'name'] = name.replace('!','')  
    
```
